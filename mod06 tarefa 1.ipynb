#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline
# %matplotlib notebook
```

```python
sinasc = pd.read_csv('SINASC_RO_2019.csv')
sinasc.drop_duplicates().shape
```

```python
sinasc = sinasc[['ESTCIVMAE', 'GESTACAO', 'CONSULTAS', 'SEXO', 'APGAR1', 'APGAR5']]
sinasc = sinasc[~sinasc['APGAR5'].isna()]
sinasc = sinasc[~sinasc['APGAR1'].isna()]
sinasc = sinasc[~sinasc['SEXO'].isna()]
sinasc = sinasc[~sinasc['GESTACAO'].isna()]
sinasc = sinasc[~sinasc['ESTCIVMAE'].isna()]
sinasc = sinasc[~sinasc['CONSULTAS'].isna()]
sinasc.head(5)
```

```python
print(sinasc.shape)
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
fig, axes = plt.subplots()
sns.countplot(ax = axes, x='ESTCIVMAE', data=sinasc)
```

```python
fig, axes = plt.subplots()
sns.countplot(ax = axes, x='APGAR1', data=sinasc)
```

```python
fig, axes = plt.subplots()
sns.countplot(ax = axes, x='APGAR5', data=sinasc)
```

```python
fig, axes = plt.subplots()
sns.countplot(ax = axes, x='SEXO', data=sinasc)
```

```python
fig, axes = plt.subplots()
sns.countplot(ax = axes, x='GESTACAO', data=sinasc)
axes.tick_params(axis="x", rotation = 30)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sinasc['apgar5_md'] = sinasc['APGAR5'].mean() /sinasc['CONSULTAS']

fig, axes = plt.subplots()
plt.rc('figure', figsize=(16, 10))
sns.countplot(ax = axes, x='apgar5_md', data=sinasc)
axes.tick_params(axis="x", rotation = 30)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, axes = plt.subplots(4, 1)
plt.rc('figure', figsize=(12, 26))

sns.countplot(ax = axes[0], x='SEXO', data=sinasc)
ax = sns.countplot(ax = axes[1], x='GESTACAO', data=sinasc)
ax.tick_params(axis='x', rotation = 45)
sns.countplot(ax = axes[2], x='ESTCIVMAE', data=sinasc)
sns.countplot(ax = axes[3], x='CONSULTAS', data=sinasc)

plt.subplots_adjust(wspace=0.3, hspace=0.7)

fig.tight_layout()
```
